## Data cleaning


```python
# we start with removing the division winnings
# we are not interested in those now

df = df[[
        #'Draw', 
        'Date', 'Winning Number 1', '2', '3', '4', '5', '6',
        'Additional Number ', 'From Last', 'Same As Day', 'Low', 'High', 'Odd',
        'Even', ' 1-10', ' 11-20', ' 21-30', ' 31-40', ' 41-50'
       # 'Division 1 Winners', 'Division 1 Prize', 'Division 2 Winners',
       # 'Division 2 Prize', 'Division 3 Winners', 'Division 3 Prize',
       # 'Division 4 Winners', 'Division 4 Prize', 'Division 5 Winners',
       # 'Division 5 Prize', 'Division 6 Winners', 'Division 6 Prize',
       # 'Division 7 Winners', 'Division 7 Prize'
        ]].copy()
df.columns
```

```python
# convert date to datetime type
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
# rename to remove spaces and standardise lowercase
df = df.rename(columns={'Date':'date', 
                        'Winning Number 1':'1',
                        'Additional Number ':'additional_number', 
                        'From Last':'from_last', 
                        'Same As Day':'same_as_day', 
                        'Low':'low', 
                        'High':'high', 
                        'Odd':'odd',
                        'Even':'even', 
                        ' 1-10':'1-10', 
                        ' 11-20':'11-20', 
                        ' 21-30':'21-30', 
                        ' 31-40':'31-40', 
                        ' 41-50':'41-50'})
df.columns
```

```python
df.isna().sum()
```

The from_last column only has a number if there is an overlap with the previous set of numbers. To remove the NA values we will convert this to 7 binary columns, corresponding to the number that was overlapped.

```python
# converting all
df['same_1'] = df.apply(lambda row: 1 if str(row['1']) in row['from_last'] else 0, axis=1)
df['same_2'] = df.apply(lambda row: 1 if str(row['2']) in row['from_last'] else 0, axis=1)
df['same_3'] = df.apply(lambda row: 1 if str(row['3']) in row['from_last'] else 0, axis=1)
df['same_4'] = df.apply(lambda row: 1 if str(row['4']) in row['from_last'] else 0, axis=1)
df['same_5'] = df.apply(lambda row: 1 if str(row['5']) in row['from_last'] else 0, axis=1)
df['same_6'] = df.apply(lambda row: 1 if str(row['6']) in row['from_last'] else 0, axis=1)
df['same_add'] = df.apply(lambda row: 1 if str(row['additional_number']) in row['from_last'] else 0, axis=1)
```
